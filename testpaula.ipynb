```python
%pip install seaborn
%pip install matplotlib
%pip install scikit-learn
%pip install pandas
%pip install numpy
%pip install openpyxl

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(data):
    return pd.read_csv(data)
```

```python

expression_df = pd.read_csv("./data/mmc2.csv", usecols = ["NK.27+11b-.BM","NK.27+11b+.BM","NK.27-11b+.BM","NK.27+11b-.Sp","NK.27+11b+.Sp","NK.27-11b+.Sp","ILC2.SI", "ILC3.NKp46-CCR6-.SI", "ILC3.NKp46+.SI", "ILC3.CCR6+.SI"])


# 2. Standardize expression values across samples
scaler = StandardScaler()
scaled_expression = scaler.fit_transform(expression_df)

# 3. Run k-means clustering (you can tune the number of clusters)
k = 15  # change depending on your dataset
kmeans = KMeans(n_clusters=k, random_state=42)
gene_clusters = kmeans.fit_predict(scaled_expression)

# 4. Add cluster labels back to the dataframe
expression_df['Cluster'] = gene_clusters

# 5. Optional: visualize average expression per cluster
cluster_mean = expression_df.groupby('Cluster').mean()

plt.figure(figsize=(10, 6))
sns.heatmap(cluster_mean, cmap="viridis")
plt.title("Average Gene Expression per Cluster")
plt.xlabel("Sample")
plt.ylabel("Gene Cluster")
plt.show()
```
