```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

expression_df = pd.read_csv("./data/mmc2.csv", usecols = ["NK.27+11b-.BM","NK.27+11b+.BM","NK.27-11b+.BM","NK.27+11b-.Sp","NK.27+11b+.Sp","NK.27-11b+.Sp","ILC2.SI", "ILC3.NKp46-CCR6-.SI", "ILC3.NKp46+.SI", "ILC3.CCR6+.SI"])
expression_df.head()
```

```python
scaler = StandardScaler()
scaled_expression = scaler.fit_transform(expression_df)
```

```python


# 3. Run k-means clustering (you can tune the number of clusters)
k = 15  # change depending on your dataset
kmeans = KMeans(n_clusters=k, random_state=42)
gene_clusters = kmeans.fit_predict(scaled_expression)

# 4. Add cluster labels back to the dataframe
expression_df['Cluster'] = gene_clusters

# 5. Optional: visualize average expression per cluster
cluster_mean = expression_df.groupby('Cluster').mean()

plt.figure(figsize=(10, 6))
sns.heatmap(cluster_mean, cmap="viridis")
plt.title("Average Gene Expression per Cluster")
plt.xlabel("Sample")
plt.ylabel("Gene Cluster")
plt.show()
```

```python
# Plotting the clusters 

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
 
#Load Data
data = pd.read_csv("./data/mmc2.csv", usecols = ["NK.27+11b-.BM","NK.27+11b+.BM","NK.27-11b+.BM","NK.27+11b-.Sp","NK.27+11b+.Sp","NK.27-11b+.Sp","ILC2.SI", "ILC3.NKp46-CCR6-.SI", "ILC3.NKp46+.SI", "ILC3.CCR6+.SI"])
pca = PCA(2)
 
#Transform the data
df = pca.fit_transform(data)
 
df.shape
```

```python
#Import required module
from sklearn.cluster import KMeans
 
#Initialize the class object
kmeans = KMeans(n_clusters= 15)
 
#predict the labels of clusters.
label = kmeans.fit_predict(df)
 
print(label)
```

```python
import matplotlib.pyplot as plt
 
#filter rows of original data
filtered_label0 = df[label == 0]
 
#plotting the results
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.show()
```

```python
#filter rows of original data
filtered_label2 = df[label == 2]
 
filtered_label8 = df[label == 8]
 
#Plotting the results
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1] , color = 'red')
plt.scatter(filtered_label8[:,0] , filtered_label8[:,1] , color = 'black')
plt.show()
```

```python
#Getting unique labels
 
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.legend()
plt.show()
```
